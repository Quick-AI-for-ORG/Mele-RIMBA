```python
import os
import time
from ultralytics import YOLO
import cv2
import numpy as np
```

```python
vidPath = 'Test/855711-hd_1920_1080_25fps.mp4'
vidoutpath ='{}_out.mp4'.format(vidPath)
cap = cv2.VideoCapture(vidPath)
ret, frame = cap.read()
h,w,_ = frame.shape
out = cv2.VideoWriter(vidoutpath,cv2.VideoWriter_fourcc(*'MP4V'), int(cap.get(cv2.CAP_PROP_FPS)),(w,h))
```

```python
modelPath = 'train41/weights/best.pt'
model = YOLO(modelPath)
threshold = 0.2
```

```python
while ret:

    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            res = f"{score:.2f} {results.names[int(class_id)].upper()}"
            cv2.putText(frame, res, (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
    
```
